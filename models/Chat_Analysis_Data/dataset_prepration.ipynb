```python
import pandas as pd
```

```python
df1 = pd.read_csv('Suicide_Detection.csv')
```

```python
df1.head()
```

```python
df1 = df1.drop(df1.columns[0], axis=1)
```

```python
df1.head()
```

```python
df1['label'] = df1['class'].map({'suicide': 'suicide', 'non-suicide': 'normal'})
```

```python
df1.head()
```

```python
df2 =pd.read_csv('cyberbullying_tweets.csv')
```

```python
df2.head()
```

```python
df2.columns = ['text', 'class']
```

```python
df2.head()
```

```python
df2['label'] = df2['class'].map({'not_cyberbullying': 'normal', 'cyberbullying': 'bullying'})
```

```python
df2.head()
```

```python
df4= pd.read_csv('labeled_data.csv ')
```

```python
df4.head()
```

```python
df4 = df4[['class','tweet']]
```

```python
df4.head()
```

```python
df4.columns= ['class','text']
```

```python
df4['label'] = df4['class'].map({1:'hate',1:'bullying',2:'hate'})
```

```python
df4.head()
```

```python
final_df = pd.concat([df1[['text','label']], df2[['text','label']], df4[['text','label']]])
```
