## Concatenating All Dataframes to create one

```python
df = pd.concat([df2, df3, df4, df5], ignore_index=True)
```

```python
df.head()
```

## Feature Engineering

#### Length Based Features

```python
df['url_length'] = df['url'].str.len()
```

long urls -> suspicious

##### Special Character counts

```python
df['dot_count'] = df['url'].str.count(r'\.')
df['dash_count'] = df['url'].str.count('-')
df['slash_count'] = df['url'].str.count('/')
df['at_count'] = df['url'].str.count('@')
```

too much symbols -> Spam/Phishing patterns

#### Ratio of digits in URLs

```python
def digit_ratio(url):
    digits = sum(c.isdigit() for c in url)
    return digits / len(url)

df['digit_ratio'] = df['url'].apply(digit_ratio)
```

greater digits ratio in URL are risky

## Checking for IP address in URL

```python
df['has_ip'] = df['url'].str.contains(r'\d+\.\d+\.\d+\.\d+').astype(int)
```

URLs having IP addresses are often malicious

#### Suspicious Keyword Detection

```python
bad_words = ['porn','xxx','sex','adult','nude','naked','boobs','breast','escort',
    'cam','cams','webcam','livecam','onlyfans','xvideo','xnxx','redtube',
    'hentai','bdsm','fetish','milf','teen','erotic','strip','stripchat','casino','bet','betting','poker','slot','jackpot','roulette',
    'blackjack','win','winnings','bonus','freebet','odds','gamble', 'free','offer','click','login','signup','verify','account',
    'secure','update','bonus','gift','promo','cheap','deal', 's3x','p0rn','pr0n','xxx','xnx','sexx','pron','sexyy']
def check_bad_words(url):
    url = str(url).lower()
    for w in bad_words:
        if w in url:
            return 1
    return 0

df['has_bad_word'] = df['url'].apply(check_bad_words)
```

#### Checking for Https protocol

```python
def check_https(url):
    url = str(url).lower()
    if 'https' in url:
        return 1
    return 0

df['has_https'] = df['url'].apply(check_https)
```

URLs usinhg https are relatively safer
